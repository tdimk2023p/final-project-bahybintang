```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load csv response form
df = pd.read_csv('../data/form_responses.csv')
df
```

```python
# Indonesia
df_indo_nswipe_wpm_mean = df.iloc[:, [7, 9, 11]].mean(axis=1)
df_indo_nswipe_acc_mean = df.iloc[:, [8, 10, 12]].mean(axis=1)
df_indo_swipe_wpm_mean = df.iloc[:, [13, 15, 17]].mean(axis=1)
df_indo_swipe_acc_mean = df.iloc[:, [14, 16, 18]].mean(axis=1)

# English
df_eng_nswipe_wpm_mean = df.iloc[:, [19, 21, 23]].mean(axis=1)
df_eng_nswipe_acc_mean = df.iloc[:, [20, 22, 24]].mean(axis=1)
df_eng_swipe_wpm_mean = df.iloc[:, [25, 27, 29]].mean(axis=1)
df_eng_swipe_acc_mean = df.iloc[:, [26, 28, 30]].mean(axis=1)

# Java
df_java_nswipe_wpm_mean = df.iloc[:, [31, 33, 35]].mean(axis=1)
df_java_nswipe_acc_mean = df.iloc[:, [32, 34, 36]].mean(axis=1)
df_java_swipe_wpm_mean = df.iloc[:, [37, 39, 41]].mean(axis=1)
df_java_swipe_acc_mean = df.iloc[:, [38, 40, 42]].mean(axis=1)
```

```python
# Boxplot for all WPM (Non-Swipe)
plt.figure()
plt.boxplot([df_indo_nswipe_wpm_mean, df_eng_nswipe_wpm_mean, df_java_nswipe_wpm_mean])
plt.xticks([1, 2, 3], ['Indonesia', 'English', 'Java'])
plt.ylabel('WPM')
plt.title('Non-Swipe')
plt.savefig('../graph/all_wpm_nswipe.png')

# Boxplot for all WPM (Swipe)
plt.figure()
plt.boxplot([df_indo_swipe_wpm_mean, df_eng_swipe_wpm_mean, df_java_swipe_wpm_mean])
plt.xticks([1, 2, 3], ['Indonesia', 'English', 'Java'])
plt.ylabel('WPM')
plt.title('Swipe')
plt.savefig('../graph/all_wpm_swipe.png')

# Boxplot for all Accuracy (Non-Swipe)
plt.figure()
plt.boxplot([df_indo_nswipe_acc_mean, df_eng_nswipe_acc_mean, df_java_nswipe_acc_mean])
plt.xticks([1, 2, 3], ['Indonesia', 'English', 'Java'])
plt.ylabel('Accuracy')
plt.title('Non-Swipe')
plt.savefig('../graph/all_acc_nswipe.png')

# Boxplot for all Accuracy (Swipe)
plt.figure()
plt.boxplot([df_indo_swipe_acc_mean, df_eng_swipe_acc_mean, df_java_swipe_acc_mean])
plt.xticks([1, 2, 3], ['Indonesia', 'English', 'Java'])
plt.ylabel('Accuracy')
plt.title('Swipe')
plt.savefig('../graph/all_acc_swipe.png')
```

```python
df_indo_nswipe_easiness = df['Bagaimana menurut Anda kemudahan menggunakan keyboard normal dalam Bahasa Indonesia'].value_counts()
df_eng_nswipe_easiness = df['Bagaimana menurut Anda kemudahan menggunakan keyboard normal dalam Bahasa Inggris?'].value_counts()
df_java_nswipe_easiness = df['Bagaimana menurut Anda kemudahan menggunakan keyboard normal dalam Bahasa Jawa?'].value_counts()

df_indo_swipe_easiness = df['Bagaimana menurut Anda kemudahan menggunakan keyboard swipe dalam Bahasa Indonesia?'].value_counts()
df_eng_swipe_easiness = df['Bagaimana menurut Anda kemudahan menggunakan keyboard swipe dalam Bahasa Inggris?'].value_counts()
df_java_swipe_easiness = df['Bagaimana menurut Anda kemudahan menggunakan keyboard swipe dalam Bahasa Jawa?'].value_counts()

df_indo_swipe_accuracy = df['Bagaimana menurut Anda akurasi keyboard swipe dalam Bahasa Indonesia?'].value_counts()
df_eng_swipe_accuracy = df['Bagaimana menurut Anda akurasi keyboard swipe dalam Bahasa Inggris?'].value_counts()
df_java_swipe_accuracy = df['Bagaimana menurut Anda akurasi keyboard swipe dalam Bahasa Jawa?'].value_counts()
```

```python
# Visualization of easiness of using keyboard (Swipe) in Indonesia, English, and Java
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Easiness of using keyboard (Swipe)')

# Easiness of using keyboard (Swipe)
axes[0].bar(df_indo_swipe_easiness.index, df_indo_swipe_easiness.values)
axes[0].set_title('Indonesia')
axes[0].set_xlabel('Easiness')
axes[0].set_ylabel('Frequency')
axes[0].set_xticks(range(1, 5))

axes[1].bar(df_eng_swipe_easiness.index, df_eng_swipe_easiness.values)
axes[1].set_title('English')
axes[1].set_xlabel('Easiness')
axes[1].set_ylabel('Frequency')
axes[1].set_xticks(range(1, 5))

axes[2].bar(df_java_swipe_easiness.index, df_java_swipe_easiness.values)
axes[2].set_title('Java')
axes[2].set_xlabel('Easiness')
axes[2].set_ylabel('Frequency')
axes[2].set_xticks(range(1, 5))

plt.tight_layout()
plt.savefig('../graph/subj_swipe_easiness.png')

# Visualization of easiness of using keyboard (Non Swipe) in Indonesia, English, and Java
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Easiness of using keyboard (Non Swipe)')

# Easiness of using keyboard (Non Swipe)
axes[0].bar(df_indo_nswipe_easiness.index, df_indo_nswipe_easiness.values)
axes[0].set_title('Indonesia')
axes[0].set_xlabel('Easiness')
axes[0].set_ylabel('Frequency')
axes[0].set_xticks(range(1, 5))

axes[1].bar(df_eng_nswipe_easiness.index, df_eng_nswipe_easiness.values)
axes[1].set_title('English')
axes[1].set_xlabel('Easiness')
axes[1].set_ylabel('Frequency')
axes[1].set_xticks(range(1, 5))

axes[2].bar(df_java_nswipe_easiness.index, df_java_nswipe_easiness.values)
axes[2].set_title('Java')
axes[2].set_xlabel('Easiness')
axes[2].set_ylabel('Frequency')
axes[2].set_xticks(range(1, 5))

plt.tight_layout()
plt.savefig('../graph/subj_nswipe_easiness.png')

# Visualization of accuracy of using keyboard (Swipe) in Indonesia, English, and Java
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Accuracy of using keyboard (Swipe)')

# Accuracy of using keyboard (Swipe)
axes[0].bar(df_indo_swipe_accuracy.index, df_indo_swipe_accuracy.values)
axes[0].set_title('Indonesia')
axes[0].set_xlabel('Accuracy')
axes[0].set_ylabel('Frequency')
axes[0].set_xticks(range(1, 5))

axes[1].bar(df_eng_swipe_accuracy.index, df_eng_swipe_accuracy.values)
axes[1].set_title('English')
axes[1].set_xlabel('Accuracy')
axes[1].set_ylabel('Frequency')
axes[1].set_xticks(range(1, 5))

axes[2].bar(df_java_swipe_accuracy.index, df_java_swipe_accuracy.values)
axes[2].set_title('Java')
axes[2].set_xlabel('Accuracy')
axes[2].set_ylabel('Frequency')
axes[2].set_xticks(range(1, 5))

plt.tight_layout()
plt.savefig('../graph/subj_swipe_accuracy.png')
```

```python
# ANOVA test for WPM (Non-Swipe)
import scipy.stats as stats
fvalue, pvalue = stats.f_oneway(df_indo_nswipe_wpm_mean, df_eng_nswipe_wpm_mean, df_java_nswipe_wpm_mean)
print('F-value:', fvalue)
print('P-value:', pvalue)

# ANOVA test for WPM (Swipe)
fvalue, pvalue = stats.f_oneway(df_indo_swipe_wpm_mean, df_eng_swipe_wpm_mean, df_java_swipe_wpm_mean)
print('F-value:', fvalue)
print('P-value:', pvalue)

# ANOVA test for Accuracy (Non-Swipe)
fvalue, pvalue = stats.f_oneway(df_indo_nswipe_acc_mean, df_eng_nswipe_acc_mean, df_java_nswipe_acc_mean)
print('F-value:', fvalue)
print('P-value:', pvalue)

# ANOVA test for Accuracy (Swipe)
fvalue, pvalue = stats.f_oneway(df_indo_swipe_acc_mean, df_eng_swipe_acc_mean, df_java_swipe_acc_mean)
print('F-value:', fvalue)
print('P-value:', pvalue)
```
