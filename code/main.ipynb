```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load csv response form
df = pd.read_csv('../data/form_responses.csv')
df
```

```python
# Indonesia
df_indo_nswipe_wpm_mean = df.iloc[:, [7, 9, 11]].mean(axis=1)
df_indo_nswipe_acc_mean = df.iloc[:, [8, 10, 12]].mean(axis=1)
df_indo_swipe_wpm_mean = df.iloc[:, [13, 15, 17]].mean(axis=1)
df_indo_swipe_acc_mean = df.iloc[:, [14, 16, 18]].mean(axis=1)

# English
df_eng_nswipe_wpm_mean = df.iloc[:, [19, 21, 23]].mean(axis=1)
df_eng_nswipe_acc_mean = df.iloc[:, [20, 22, 24]].mean(axis=1)
df_eng_swipe_wpm_mean = df.iloc[:, [25, 27, 29]].mean(axis=1)
df_eng_swipe_acc_mean = df.iloc[:, [26, 28, 30]].mean(axis=1)

# Java
df_java_nswipe_wpm_mean = df.iloc[:, [31, 33, 35]].mean(axis=1)
df_java_nswipe_acc_mean = df.iloc[:, [32, 34, 36]].mean(axis=1)
df_java_swipe_wpm_mean = df.iloc[:, [37, 39, 41]].mean(axis=1)
df_java_swipe_acc_mean = df.iloc[:, [38, 40, 42]].mean(axis=1)
```

```python
# Boxplot for all WPM (Non-Swipe)
plt.figure()
plt.boxplot([df_indo_nswipe_wpm_mean, df_eng_nswipe_wpm_mean, df_java_nswipe_wpm_mean])
plt.xticks([1, 2, 3], ['Indonesia', 'English', 'Java'])
plt.title('Non-Swipe')
plt.savefig('../data/all_wpm_nswipe.png')

# Boxplot for all WPM (Swipe)
plt.figure()
plt.boxplot([df_indo_swipe_wpm_mean, df_eng_swipe_wpm_mean, df_java_swipe_wpm_mean])
plt.xticks([1, 2, 3], ['Indonesia', 'English', 'Java'])
plt.title('Swipe')
plt.savefig('../data/all_wpm_swipe.png')

# Boxplot for all Accuracy (Non-Swipe)
plt.figure()
plt.boxplot([df_indo_nswipe_acc_mean, df_eng_nswipe_acc_mean, df_java_nswipe_acc_mean])
plt.xticks([1, 2, 3], ['Indonesia', 'English', 'Java'])
plt.title('Non-Swipe')
plt.savefig('../data/all_acc_nswipe.png')

# Boxplot for all Accuracy (Swipe)
plt.figure()
plt.boxplot([df_indo_swipe_acc_mean, df_eng_swipe_acc_mean, df_java_swipe_acc_mean])
plt.xticks([1, 2, 3], ['Indonesia', 'English', 'Java'])
plt.title('Swipe')
plt.savefig('../data/all_acc_swipe.png')
```

```python
df_gender = df['Jenis Kelamin']
df_frequency = df['Frekuensi penggunaan smartphone dalam sehari']
```

```python
# Correlation between frequency and WPM (Non-Swipe)
plt.figure()
plt.scatter(df_frequency, df_indo_nswipe_wpm_mean, label='Indonesia')
plt.scatter(df_frequency, df_eng_nswipe_wpm_mean, label='English')
plt.scatter(df_frequency, df_java_nswipe_wpm_mean, label='Java')
plt.legend()
plt.xlabel('Frequency')
plt.ylabel('WPM')
plt.title('Non-Swipe')
plt.savefig('../data/frequency_wpm_nswipe.png')

# Correlation between frequency and WPM (Swipe)
plt.figure()
plt.scatter(df_frequency, df_indo_swipe_wpm_mean, label='Indonesia')
plt.scatter(df_frequency, df_eng_swipe_wpm_mean, label='English')
plt.scatter(df_frequency, df_java_swipe_wpm_mean, label='Java')
plt.legend()
plt.xlabel('Frequency')
plt.ylabel('WPM')
plt.title('Swipe')
plt.savefig('../data/frequency_wpm_swipe.png')

# Correlation between frequency and Accuracy (Non-Swipe)
plt.figure()
plt.scatter(df_frequency, df_indo_nswipe_acc_mean, label='Indonesia')
plt.scatter(df_frequency, df_eng_nswipe_acc_mean, label='English')
plt.scatter(df_frequency, df_java_nswipe_acc_mean, label='Java')
plt.legend()
plt.xlabel('Frequency')
plt.ylabel('Accuracy')
plt.title('Non-Swipe')
plt.savefig('../data/frequency_acc_nswipe.png')

# Correlation between frequency and Accuracy (Swipe)
plt.figure()
plt.scatter(df_frequency, df_indo_swipe_acc_mean, label='Indonesia')
plt.scatter(df_frequency, df_eng_swipe_acc_mean, label='English')
plt.scatter(df_frequency, df_java_swipe_acc_mean, label='Java')
plt.legend()
plt.xlabel('Frequency')
plt.ylabel('Accuracy')
plt.title('Swipe')
plt.savefig('../data/frequency_acc_swipe.png')
```
